# **Task 1: Dataset Creation**


```python
X = [
    [2, 40], [5, 90], [1, 30], [10, 85], [7, 70],
    [5, 60], [9, 95], [4, 60], [6, 80], [2, 10],
    [8, 100], [5, 45], [1, 10], [2, 20]
]

# Rule: Study Hours * Attendance > 250, they Pass.
Y = [0, 1, 0, 1, 1, 1, 1, 0, 1, 0, 1, 0, 0, 0]
```

# **Task 2 & 3:Perceptron Initialization & Activation Function**


```python
# Initializing with small fixed values as starting small prevents massive errors early on.
w1 = 0.1
w2 = 0.1
bias = -0.5
learning_rate = 0.001  # Small steps for accuracy

def activation(weighted_sum):
    # Threshold function: 1 if positive, 0 if negative
    if weighted_sum > 0:
        return 1
    else:
        return 0
```

# **Task 4ï¸: Training Loop**

```python
epochs = 30

for epoch in range(epochs):
    total_errors = 0

    for i in range(len(X)):
        xi = X[i]
        yi = Y[i]
        z = (xi[0] * w1) + (xi[1] * w2) + bias

        # Applying the threshold (Activation Function)
        prediction = 1 if z >= 0 else 0

        error = yi - prediction

        # We only update if there is an error (misclassification)
        if error != 0:
            w1 = w1 + (learning_rate * error * xi[0])
            w2 = w2 + (learning_rate * error * xi[1])
            bias = bias + (learning_rate * error)
            total_errors += 1

    print(f"Epoch {epoch+1} finished. Total Misclassified: {total_errors}")
    if total_errors == 0:
        print("Model has perfectly learned the dataset!")
        break
```

# **Task 5: User Input Testing**

```python
# Taking user input
hours_input = float(input("Enter study hours: "))
attendance_input = float(input("Enter attendance %: "))

final_sum = (hours_input * w1) + (attendance_input * w2) + bias
result = activation(final_sum)

if result == 1:
    print("The student is likely to Pass")
else:
    print("The student is likely to Fail")
```

**1. How the Dataset was Created**

The dataset consists of 14 unique data points representing student performance. Each point includes two features: Study Hours and Attendance Percentage. To assign the labels (Pass/Fail), I used a custom multiplication rule: if the product of Study Hours and Attendance is greater than 250, the student is labeled as Pass (1); otherwise, they are labeled as Fail (0). This rule ensures that a student needs a balance of both effort and presence to succeed, which creates a clear pattern for the Perceptron to learn

**2. Why the Learning Rate was Chosen**

I chose a learning rate of 0.001. During early testing, I noticed that the input values (especially attendance which goes up to 100) are relatively large. A higher learning rate caused the weights to jump too drastically, making the model "overshoot" the ideal values

**3. How the Model was Verified**

To verify that the model was actually learning, I monitored two specific indicators:

**Error Reduction:** I printed the total number of misclassifications at the end of each epoch. I observed that the error count consistently decreased as the epochs progressed, proving the weights were adjusting correctly.


**User Input Testing:** After training was finished, I used the input() function to test the model with new values. The model correctly predicted the outcome based on my original rule, confirming it had successfully mapped the relationship between the inputs and the labels.
