```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns

scalar = 3.7
arr = np.array([10.2,20.6,30.5,40.4,50.7,60.9])

# Trigonometric functions
sin_values = np.sin(arr)
cos_values = np.cos(arr)
tan_values = np.tan(arr)
asin_values = np.arcsin(np.clip(arr / np.max(arr), -1, 1))  # Normalized and Clipping for valid range [-1, 1]
acos_values = np.arccos(np.clip(arr / np.max(arr), -1, 1))
atan_values = np.arctan(arr)

exp_values = np.exp(arr)
log_values = np.log(arr)  # Natural logarithm

# Absolute value and square root
abs_values = np.abs(arr)
sqrt_values = np.sqrt(arr)

# Remainder when divided by a scalar (e.g., 2)
rem_values = np.remainder(arr, 2)

# Rounding functions
round_values = np.round(arr)
floor_values = np.floor(arr)
ceil_values = np.ceil(arr)

# Display results
print("sin:", sin_values)
print("cos:", cos_values)
print("tan:", tan_values)
print("asin:", asin_values)
print("acos:", acos_values)
print("atan:", atan_values)
print("exp:", exp_values)
print("log:", log_values)
print("abs:", abs_values)
print("sqrt:", sqrt_values)
print("rem:", rem_values)
print("round:", round_values)
print("floor:", floor_values)
print("ceil:", ceil_values)
```

```python
import numpy as np

matrix_2x3_float = np.array([[1.1, 2.5, 3.2], [4.8, 5.3, 6.9]])

# 1. Maximum value and index of max element
max_value = np.max(matrix_2x3_float)
max_index_flat = np.argmax(matrix_2x3_float)
max_index_2d = np.unravel_index(max_index_flat, matrix_2x3_float.shape)

# 2. Minimum value and index of min element
min_value = np.min(matrix_2x3_float)
min_index_flat = np.argmin(matrix_2x3_float)
min_index_2d = np.unravel_index(min_index_flat, matrix_2x3_float.shape)

# 3. Length of the matrix (number of elements)
matrix_length = matrix_2x3_float.size

# 4. Sorting
# Ascending sort (default)
sorted_matrix_flat_asc = np.sort(matrix_2x3_float, axis=None) # Sorts flattened array
sorted_matrix_rows_asc = np.sort(matrix_2x3_float, axis=1) # Sorts along rows

# Descending sort
sorted_matrix_flat_desc = np.sort(matrix_2x3_float, axis=None)[::-1] # Sorts flattened array in descending order
sorted_matrix_rows_desc = -np.sort(-matrix_2x3_float, axis=1) # Sorts along rows in descending order


# 5. Sum of elements
sum_values = np.sum(matrix_2x3_float)

# 6. Product of elements
prod_values = np.prod(matrix_2x3_float)

# 7. Median value
median_value = np.median(matrix_2x3_float)

# 8. Mean value
mean_value = np.mean(matrix_2x3_float)

# 9. Standard deviation
std_dev = np.std(matrix_2x3_float)

# Display results
print("Matrix (float):\n", matrix_2x3_float)
print("\nMax value:", max_value, "at flattened index", max_index_flat, "and 2D index", max_index_2d)
print("Min value:", min_value, "at flattened index", min_index_flat, "and 2D index", min_index_2d)
print("Length of matrix (number of elements):", matrix_length)

print("\nSorted flattened matrix (ascending):", sorted_matrix_flat_asc)
print("Sorted flattened matrix (descending):", sorted_matrix_flat_desc)
print("\nSorted matrix along rows (ascending):\n", sorted_matrix_rows_asc)
print("Sorted matrix along rows (descending):\n", sorted_matrix_rows_desc)

print("\nSum of elements:", sum_values)
print("Product of elements:", prod_values)
print("Median value:", median_value)
print("Mean value:", mean_value)
print("Standard deviation:", std_dev)
```

```python

import matplotlib.pyplot as plt
import numpy as np
# Replace static matrix with random numbers
random_matrix = np.random.uniform(low=0, high=100, size=(4, 5)) # Example: 4x5 matrix with values between 0 and 100

# Visualize the matrix using seaborn.heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(random_matrix, annot=True, cmap="viridis", fmt=".2f")
plt.title("Heatmap of Random Matrix")
plt.xlabel("Column Index")
plt.ylabel("Row Index")
plt.show()

# You can also visualize using matplotlib's imshow (less common for simple value visualization)
plt.figure(figsize=(8, 6))
plt.imshow(random_matrix, cmap="viridis")
plt.colorbar(label="Value")
plt.title("Matplotlib Visualization of Random Matrix")
plt.xlabel("Column Index")
plt.ylabel("Row Index")
plt.show()
```
